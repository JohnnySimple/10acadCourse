## TASK 1 - User Overview Analysis

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('challenge2-data.csv')
data = pd.DataFrame(data)
m = pd.notnull(data['MSISDN/Number'])
data = data[m]
data.shape
```

```python
data.T
```

```python
# top 10 handsets used by customers
top_handsets = data['Handset Type'].value_counts().head(10)
print('\n\nTop 10 handsets used by customers\n\n', top_handsets)

# top 3 handset manufacturers
top_manu = data['Handset Manufacturer'].value_counts().head(3)
print('\n\nTop 3 handset manufacturers\n\n', top_manu)
```

```python
# top 5 handsets per Apple
top_apple_handsets = data['Handset Type'][data['Handset Manufacturer'] == 'Apple'].value_counts().head()
print('\n\nTop 5 handsets of Apple\n\n', top_apple_handsets)

# top 5 handsets per Samsung
top_samsung_handsets = data['Handset Type'][data['Handset Manufacturer'] == 'Samsung'].value_counts().head()
print('\n\nTop 5 handsets of Samsung\n\n', top_samsung_handsets)

# top 5 handsets per Huawei
top_huawei_handsets = data['Handset Type'][data['Handset Manufacturer'] == 'Huawei'].value_counts().head()
print('\n\nTop 5 handsets of Huawei\n\n', top_huawei_handsets)
```

## Task 1.1

```python
# Number of xDR sessions per user
print('Showing the number of xDR sessions per user')
xdr_sessions = data.groupby('MSISDN/Number').count()['Bearer Id']
xdr_sessions = pd.DataFrame(xdr_sessions)
xdr_sessions.columns = ['xDR Sessions']
xdr_sessions
```

```python
# Session duration per user
print('Showing session duration per user')
s_duration = data.groupby('MSISDN/Number').sum()['Dur. (ms)']
s_duration = pd.DataFrame(s_duration)
s_duration.columns = ['Session_Duration']
s_duration
```

```python
# Total download and upload data per user
print('Showing total download and upload data per user')
total_ul = data.groupby('MSISDN/Number').sum()['Total UL (Bytes)']
total_dl = data.groupby('MSISDN/Number').sum()['Total DL (Bytes)']
total_dlul = pd.DataFrame(total_ul)
total_dlul['Total DL (Bytes)'] = total_dl
total_dlul['Total'] = total_dlul['Total UL (Bytes)'] + total_dlul['Total DL (Bytes)']
total_dlul
```

```python
total_vol = pd.DataFrame()
total_vol['MSISDN/Number'] = data['MSISDN/Number']
total_vol['Social Media(Bytes)'] = data['Social Media DL (Bytes)'] + data['Social Media UL (Bytes)']
total_vol['Google(Bytes)'] = data['Google DL (Bytes)'] + data['Google UL (Bytes)']
total_vol['Email(Bytes)'] = data['Email DL (Bytes)'] + data['Email UL (Bytes)']
total_vol['Youtube(Bytes)'] = data['Youtube DL (Bytes)'] + data['Youtube UL (Bytes)']
total_vol['Netflix(Bytes)'] = data['Netflix DL (Bytes)'] + data['Netflix UL (Bytes)']
total_vol['Gaming(Bytes)'] = data['Gaming DL (Bytes)'] + data['Gaming UL (Bytes)']
total_vol['Other(Bytes)'] = data['Other DL (Bytes)'] + data['Other UL (Bytes)']
print('Showing total data volume of various applications used by users.')
total_vol
```
